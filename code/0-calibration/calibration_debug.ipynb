```python
import imageio

frame = imageio.imread('/mnt/f/Github/MARMOT/data/0-calibration/opensfm/images/cam360_1200.jpg')
```

```python
import py360convert

frame_cubemap = py360convert.e2c(frame, face_w=512, mode='bilinear', cube_format='list')

import matplotlib.pyplot as plt
for i in range(len(frame_cubemap)):
    plt.subplot(2,3,i+1)
    plt.imshow(frame_cubemap[i])
    # save frames to disk
    imageio.imwrite(f'/mnt/f/Github/MARMOT/data/0-calibration/opensfm/images/cam360_1200_{i}.jpg', frame_cubemap[i])
```
